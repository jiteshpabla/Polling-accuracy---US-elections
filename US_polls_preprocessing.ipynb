# actual vote data

```
path_president_candidate = 'president_county_candidate.csv'

usa_election = pd.read_csv(path_president_candidate)

usa_election
```

```
DEM_election = usa_election[usa_election['party']=='DEM']
DEM_election = DEM_election.groupby('state').sum()
DEM_election.rename(columns={"total_votes": "DEM_votes", "won": "DEM_won_counties"}, inplace=True)
DEM_election.head()
```

```
REP_election = usa_election[usa_election['party']=='REP']
REP_election = REP_election.groupby('state').sum()
REP_election.rename(columns={"total_votes": "REP_votes", "won": "REP_won_counties"}, inplace=True)
REP_election.head()
```

```
notREP = usa_election[usa_election['party']!='REP']
OTH_election = notREP[notREP['party']!='DEM']
OTH_election = OTH_election.groupby('state').sum()
OTH_election.rename(columns={"total_votes": "OTH_votes", "won": "OTH_won_counties"}, inplace=True)
OTH_election.head()
```

```
president_election = pd.concat([REP_election, DEM_election, OTH_election], axis=1)
president_election.head()
```

```
president_election['winner'] = np.where(president_election['REP_votes'] > president_election['DEM_votes'], 'Donald Trump', 'Joe Biden')
president_election.reset_index(inplace=True)
president_election
```

```
share_rep = np.zeros(len(president_election))
share_dem = np.zeros(len(president_election))
share_oth = np.zeros(len(president_election))
for idx, row in president_election.iterrows():
  #print(row)
  total = row["REP_votes"] + row["DEM_votes"] + row["OTH_votes"]
  share_rep[idx] = row["REP_votes"]/total *100
  share_dem[idx] = row["DEM_votes"]/total *100
  share_oth[idx] = row["OTH_votes"]/total *100

president_election["share_REP"] = share_rep
president_election["share_DEM"] = share_dem
president_election["share_OTH"] = share_oth

president_election
```

# combiniing the 2

```
result = final.merge(president_election, on="state")
result
```

```
error_rep = np.zeros(len(result))
error_dem = np.zeros(len(result))
error_oth = np.zeros(len(result))
error_tot = np.zeros(len(result))
for idx, row in result.iterrows():
  #print(row)
  error_rep[idx] = row["share_REP"] - row["voteshare_inc"]
  error_dem[idx] = row["share_DEM"] - row["voteshare_chal"]
  error_oth[idx] = row["share_OTH"] - row["voteshare_other"]
  error_tot[idx] = abs(row["share_REP"] - row["voteshare_inc"]) + abs(row["share_REP"] - row["voteshare_inc"]) + abs(row["share_OTH"] - row["voteshare_other"])

result["error_REP"] = error_rep
result["error_DEM"] = error_dem
result["error_OTH"] = error_oth
result["error_TOT"] = error_tot

result
```

```
result['projected_winner'] = np.where(result['winstate_inc'] > result['winstate_chal'], 'Donald Trump', 'Joe Biden')
result['right_wrong'] = np.where(result['winner'] == result['projected_winner'], 1, 0)
```
